```python
import pandas as pd
import numpy as np

fp = pd.read_csv("train_data.csv")
fn = pd.read_csv("test_data.csv")

#training data and test data標準化
#把一些不合理的資料換成平均(有扣掉不合理的資料)
list_P = [] #Pregnancies
test_P = []
for i in fp["Pregnancies"]:
    list_P.append((i-fp["Pregnancies"].min())/fp["Pregnancies"].max())
for i in fn["Pregnancies"]:
    test_P.append((i-fp["Pregnancies"].min())/fp["Pregnancies"].max())
    
list_G = [] #Glucose
test_G = [] 
for i in fp["Glucose"]:
    if i <= 0:
        list_G.append((fp["Glucose"][fp["Glucose"]>0].mean()-fp["Glucose"][fp["Glucose"]>0].min())/fp["Glucose"].max())
    else:
        list_G.append((i-fp["Glucose"][fp["Glucose"]>0].min())/fp["Glucose"].max())
for i in fn["Glucose"]:
    if i <= 0:
        test_G.append((fp["Glucose"][fp["Glucose"]>0].mean()-fp["Glucose"][fp["Glucose"]>0].min())/fp["Glucose"].max())
    else:
        test_G.append((i-fp["Glucose"][fp["Glucose"]>0].min())/fp["Glucose"].max())
        
list_B = [] #BloodPressure
test_B = []
for i in fp["BloodPressure"]:
    if i <= 0:
        list_B.append((fp["BloodPressure"][fp["BloodPressure"]>0].mean()-fp["BloodPressure"][fp["BloodPressure"]>0].min())/fp["BloodPressure"].max())
    else:
        list_B.append((i-fp["BloodPressure"][fp["BloodPressure"]>0].min())/fp["BloodPressure"].max())
for i in fn["BloodPressure"]:
    if i <= 0:
        test_B.append((fp["BloodPressure"][fp["BloodPressure"]>0].mean()-fp["BloodPressure"][fp["BloodPressure"]>0].min())/fp["BloodPressure"].max())
    else:
        test_B.append((i-fp["BloodPressure"][fp["BloodPressure"]>0].min())/fp["BloodPressure"].max())

list_S = [] #SkinThickness
test_S = []
for i in fp["SkinThickness"]:
    if i <= 0:
        list_S.append((fp["SkinThickness"][fp["SkinThickness"]>0].mean()-fp["SkinThickness"][fp["SkinThickness"]>0].min())/fp["SkinThickness"].max())
    else:
        list_S.append((i-fp["SkinThickness"][fp["SkinThickness"]>0].min())/fp["SkinThickness"].max())
for i in fn["SkinThickness"]:
    if i <= 0:
        test_S.append((fp["SkinThickness"][fp["SkinThickness"]>0].mean()-fp["SkinThickness"][fp["SkinThickness"]>0].min())/fp["SkinThickness"].max())
    else:
        test_S.append((i-fp["SkinThickness"][fp["SkinThickness"]>0].min())/fp["SkinThickness"].max())
        
list_I = [] #Insulin
test_I = []
for i in fp["Insulin"]:
    if i <= 0:
        list_I.append((fp["Insulin"][fp["Insulin"]>0].mean()-fp["Insulin"][fp["Insulin"]>0].min())/fp["Insulin"].max())
    else:
        list_I.append((i-fp["Insulin"][fp["Insulin"]>0].min())/fp["Insulin"].max())
for i in fn["Insulin"]:
    if i <= 0:
        test_I.append((fp["Insulin"][fp["Insulin"]>0].mean()-fp["Insulin"][fp["Insulin"]>0].min())/fp["Insulin"].max())
    else:
        test_I.append((i-fp["Insulin"][fp["Insulin"]>0].min())/fp["Insulin"].max())        

list_BMI = [] #BMI
test_BMI = []
for i in fp["BMI"]:
    if i <= 0:
        list_BMI.append((fp["BMI"][fp["BMI"]>0].mean()-fp["BMI"][fp["BMI"]>0].min())/fp["BMI"].max())
    else:
        list_BMI.append((i-fp["BMI"][fp["BMI"]>0].min())/fp["BMI"].max())
for i in fn["BMI"]:
    if i <= 0:
        test_BMI.append((fp["BMI"][fp["BMI"]>0].mean()-fp["BMI"][fp["BMI"]>0].min())/fp["BMI"].max())
    else:
        test_BMI.append((i-fp["BMI"][fp["BMI"]>0].min())/fp["BMI"].max())
        
list_D = [] #DiabetesPedigreeFunction
test_D = []
for i in fp["DiabetesPedigreeFunction"]:
    if i <= 0:
        list_D.append((fp["DiabetesPedigreeFunction"][fp["DiabetesPedigreeFunction"]>0].mean()-fp["DiabetesPedigreeFunction"][fp["DiabetesPedigreeFunction"]>0].min())/fp["DiabetesPedigreeFunction"].max())
    else:
        list_D.append((i-fp["DiabetesPedigreeFunction"][fp["DiabetesPedigreeFunction"]>0].min())/fp["DiabetesPedigreeFunction"].max())
for i in fn["DiabetesPedigreeFunction"]:
    if i <= 0:
        test_D.append((fp["DiabetesPedigreeFunction"][fp["DiabetesPedigreeFunction"]>0].mean()-fp["DiabetesPedigreeFunction"][fp["DiabetesPedigreeFunction"]>0].min())/fp["DiabetesPedigreeFunction"].max())
    else:
        test_D.append((i-fp["DiabetesPedigreeFunction"][fp["DiabetesPedigreeFunction"]>0].min())/fp["DiabetesPedigreeFunction"].max())

        
list_Age = [] #Age
test_Age = []
for i in fp["Age"]:
    if i <= 0:
        list_Age.append((fp["Age"][fp["Age"]>0].mean()-fp["Age"][fp["Age"]>0].min())/fp["Age"].max())
    else:
        list_Age.append((i-fp["Age"][fp["Age"]>0].min())/fp["Age"].max())
for i in fn["Age"]:
    if i <= 0:
        test_Age.append((fp["Age"][fp["Age"]>0].mean()-fp["Age"][fp["Age"]>0].min())/fp["Age"].max())
    else:
        test_Age.append((i-fp["Age"][fp["Age"]>0].min())/fp["Age"].max())


#歐幾里得距離

def distance(x,y):
    return (x-y)**2

train_outcome = fp["Outcome"].values #get training data outcome
test_outcome = fn["Outcome"].values #get test data outcome
def cal(x,y): #看k=5的鄰居，是哪類
    one = 0
    zero = 0
    for i in x:
        if y[i[0]] == 1:
            one+=1
        else:
            zero += 1
    if one > zero:
        return 1
    else:
        return 0
    
predict_out = []    
for i in range(len(fn)): #先跑test data
    dis = dict()
    for j in range(len(fp)): #在跑training data
        a = distance(list_P[j],test_P[i])
        b = distance(list_G[j],test_G[i])
        c = distance(list_B[j],test_B[i])
        d = distance(list_S[j],test_S[i])
        e = distance(list_I[j],test_I[i])
        f = distance(list_BMI[j],test_BMI[i])
        g = distance(list_D[j],test_D[i])
        h = distance(list_Age[j],test_Age[i])
        
        Dis =(a+b+c+d+e+f+g+h)**0.5
        dis[j] = Dis

    dis = sorted(dis.items(), key = lambda x:x[1]) #排序
    dis = dis[0:5] #取前五個 ， k= 5

    out = cal(dis,train_outcome)
    predict_out.append(out)
  

def accuracy(predict_out,test_outcome):
    count =0
    for i in range(len(test_outcome)): #run all data
        if predict_out[i] == test_outcome[i]: #遇到這個元素時，兩邊一樣
            count+=1
    return count/len(test_outcome)

def recall(predict_out,test_outcome):
    count1 = 0 #分母
    count2 = 0 #分子
    for i in range(len(predict_out)): #run all data
        if predict_out[i] == 1:
            count1 +=1
            if test_outcome[i] == 1:
                count2 +=1
    return count2/count1

print("Accuracy rate is :",accuracy(predict_out,test_outcome))
print("Recall rate is :",recall(predict_out,test_outcome))
```
